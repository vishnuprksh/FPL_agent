```python
import pandas as pd
import requests

# Fetch data from the Fantasy Premier League API
url = "https://fantasy.premierleague.com/api/bootstrap-static/"
response = requests.get(url)
data = response.json()

# Extract elements (players) data
elements = data['elements']

# Create DataFrame from elements data
df = pd.DataFrame(elements)

# Display the DataFrame
print(f"Total players: {len(df)}")
df.head()
```

```python
# Check for null values in ep_this and ep_next columns
print("Null values in ep_this:", df['ep_this'].isnull().sum())
print("Null values in ep_next:", df['ep_next'].isnull().sum())

# Show rows with null values in either column
null_ep_this = df[df['ep_this'].isnull()]
null_ep_next = df[df['ep_next'].isnull()]

print(f"\nRows with null ep_this: {len(null_ep_this)}")
print(f"Rows with null ep_next: {len(null_ep_next)}")

if len(null_ep_this) > 0:
    print("\nSample rows with null ep_this:")
    print(null_ep_this[['id', 'first_name', 'second_name', 'ep_this', 'ep_next']].head())

if len(null_ep_next) > 0:
    print("\nSample rows with null ep_next:")
    print(null_ep_next[['id', 'first_name', 'second_name', 'ep_this', 'ep_next']].head())
```
